## Scalling of Data Set

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range = (0, 1))
data_training_scaled = scaler.fit_transform(training_set)
```

```python
# Extracts sequences of 60 time steps as features and their corresponding labels from scaled training data.
features_set = []
labels = []
for i in range(60, 586):
  features_set.append(data_training_scaled[i - 60:i, 0])
  labels.append(data_training_scaled[i, 0])
```

```python
features_set, labels = np.array(features_set), np.array(labels)
```

```python
features_set = np.reshape(features_set, (features_set.shape[0], features_set.shape[1], 1))
features_set.shape
```

##LSTM

```python
!pip install tensorflow
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LSTM
```

```python
model = Sequential()
```

```python
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
# Fit the model using features_set as input and labels as target output
model.fit(features_set, labels, epochs=50, batch_size=20)
```
